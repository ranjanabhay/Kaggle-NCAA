```python
import pandas as pd
import numpy as np
```

```python
games_df = pd.read_csv('./DataFiles/NCAATourneyCompactResults.csv')
games_df['SecondaryTourney'] = 'NCAA'
games_df.head(n=5)
```

```python
regular_df = pd.read_csv('./DataFiles/RegularSeasonCompactResults.csv')
regular_df['SecondaryTourney'] = 'Regular'
regular_df.head(n=5)
```

```python
# Reading the SecondaryTourney Compact result
secondary_df = pd.read_csv('./DataFiles/SecondaryTourneyCompactResults.csv')
secondary_df.head(n=5)
```

```python
games_df = pd.concat([games_df,secondary_df,regular_df],ignore_index=True,axis=0)
```

```python
games_df.shape[0]
```

```python
# Mapping the tournament with a numerical value.
tournament_dictionary = {'CBI':1,'CIT':2,'NIT':3,'V16':4,'NCAA':5,'Regular':6}
```

```python
WLoc ={'A':1,'H':2,'N':3}
```

```python
games_df['SecondaryTourney'] = games_df['SecondaryTourney'].map(tournament_dictionary)
games_df['SecondaryTourney'].value_counts()
```

```python
games_df['WLoc'] = games_df['WLoc'].map(WLoc)
```

```python
#games_df['Game_ID'] = games_df.apply(lambda x: '_'.join(map(str,[x['Season']]+sorted([x['WTeamID'],x['LTeamID']]))),axis=1)
games_df.head(n=3)
```

```python
# Create Team IDS
games_df['ID'] = games_df.apply(lambda x:'_'.join(map(str,[x['Season']]+sorted([x['WTeamID'],x['LTeamID']]))),axis=1)
games_df['T1'] = games_df.apply(lambda x: sorted([x['WTeamID'],x['LTeamID']])[0], axis = 1)
games_df['T2'] = games_df.apply(lambda x: sorted([x['WTeamID'],x['LTeamID']])[1], axis = 1)
games_df['T1_T2'] = games_df.apply(lambda x: '_'.join(map(str,sorted([x['WTeamID'],x['LTeamID']]))),axis=1)
games_df['Season_T1'] = games_df.apply(lambda x: '_'.join(map(str,[x['Season'],x['T1']])),axis=1)
games_df['Season_T2'] = games_df.apply(lambda x: '_'.join(map(str,[x['Season'],x['T2']])),axis=1)
games_df['Score_Diff'] = games_df['WScore'] - games_df['LScore']
games_df['Score_Diff_Percent'] = (games_df['WScore'] - games_df['LScore'])/(games_df['WScore'] + games_df['LScore'])
games_df.head(n=5)
```

games_df[game_t1_id] = games_df.apply(lambda x: sorted([x['WteamID'],x['LTeamID']]), axis = 1)
games_df.head(n=5)

```python
games_df['Pred'] = games_df.apply(lambda x: 1 if sorted([x['WTeamID'],x['LTeamID']])[0] == x['WTeamID'] else 0 , axis=1)
games_df['Score_Diff_Norm'] = games_df.apply(lambda x: x['Score_Diff']* -1 if x['Pred']==0 else x['Score_Diff'], axis=1)
games_df['Score_Diff_Percent_Norm'] = games_df.apply(lambda x: x['Score_Diff_Percent']* -1 if x['Pred']==0 else x['Score_Diff'], axis=1)
games_df.head(n=20)
```
